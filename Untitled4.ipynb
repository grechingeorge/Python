В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность.

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

### Предобработка данных
**2.1 Обработка пропусков и неявных дубликатов**

```python
data.isna().sum()
```

**Анализ пропусков:** В датафрейме имеется большое кол-во пропусков в столбцах ceilling_height, is_apartment, balcony. Можно предположить, что пропуски в стобце ceilling_height означают стандартвую стреднюю высоту потолков. Пропуски is_apartment можно заменить на False, тк статистически квартир больше, чем апартаментов. Пропуски balcony возникли из-за того, что балкон впринципе отсутствует.
Пропуски в столбцах, заполненных пользователями, можно объяснить тем, незнанием пользователей расстояния до цента/парков/аэропорта
Пропуски в days_exposition могут означать, что объяеление все еще активно

```python
#заполняем пропуски в столбце balcony, скорей всего если стоит пустое значение то значит, это значит что балконов 0
good_data = data
good_data['balcony'] = good_data['balcony'].fillna(0)
good_data['is_apartment'] = good_data['is_apartment'].fillna(False)

#удалим строки с пустыми значениями в столбцах 'living_area' 'floors_total'
good_data = good_data.dropna(subset=['floors_total'])

#расчитаем долю жилой площади от общей площади и заполним пропуски в living_area

living_area_part = (good_data['total_area'] / good_data['living_area']).median()
good_data['living_area'].fillna(round(good_data['total_area'] / living_area_part, 2), inplace=True)
```

```python
#избавимся от неявных дупликатов в столбце locality_name

good_data['locality_name'].unique()
good_data['locality_name'] = (
    good_data['locality_name']
    .str.replace('ё', 'е')
    .replace(['городской поселок', 'поселок городского типа'], 'поселок', regex=True)
)

good_data.duplicated().sum()
good_data['locality_name'].nunique()
```

**Вывод 2.1** Были заполены пропуски в is_apartmen по принципу, если пропуск - значит это не апартаменты. Пропуски в  balcony были заполнены нулями. Пропускам living_area, floors_total, kitchen_area нет логического объяснения, поэтому эти данные были удалены из ДФ
Также были приведены к одинаковому названия наименования посёлок/поселок городского типа и тд.

Данные с пропусками, введеные пользователем, были не тронуты, тк непонятна причина отсутствия значений

**2.2 Обработка аномальных значений**

```python
good_data.describe()
```

```python
#предположим, что значения потолков > 20.0, был написан неправильно десятичный знак,
#поэтому разделим такие значения на 10
good_data.loc[good_data['ceiling_height'] >= 20, 'ceiling_height'] /= 10

#заполним пропуски высоты потолоков медианным значением
good_data.loc[good_data['ceiling_height'].isna(), 'ceiling_height'] = good_data['ceiling_height'].mean()

good_data = good_data.reset_index(drop=True)
good_data.isna().sum()
```

```python
#изменим тип данных
good_data = good_data.astype({'balcony': 'int64', 'floors_total': 'int64'})
```

**Вывод 2.2** Для изменения типа данных, сначала были заполнены пропуски в столбце balcony и в нем же исправлены аномальные значения. Столбцы balcony, floors_total имели тип данных float, хотя очевидно, что эти параметры могут иметь только целочиссленный тип данных

### Посчитаем и добавим в таблицу новые столбцы

```python
#добавим столбец - цена 1кв.м.
good_data['price_for_1m'] = round(good_data['last_price'] / good_data['total_area'], 2)

#добавление столбцов с днем недели, месяцем и годов размещения объявления
good_data['first_day_exposition'] = pd.to_datetime(good_data['first_day_exposition'])
good_data['day_exposition'] = good_data['first_day_exposition'].dt.weekday
good_data['month_exposition'] = pd.DatetimeIndex(good_data['first_day_exposition']).month
good_data['year_exposition'] = pd.DatetimeIndex(good_data['first_day_exposition']).year

#расстояние до центра города в км
good_data['city_center_km'] = round(good_data['cityCenters_nearest'] / 1000, 2)



#добавит столбец с типом этажа
def type_floor(floor):
    try:
        if floor == 1:
            return 'Первый'
        else:
            return 'Другой'
    except:
        pass

good_data['type_floor'] = good_data['floor'].apply(type_floor)
good_data.loc[good_data['floor'] == good_data['floors_total'], 'type_floor'] = 'Последний'
```

**4.5 Средняя цена каждого километра**

```python
#удалим пропуски в столбце city_center_km и вычислим среднюю стоимость одного километра
data_spb = (
    good_data[good_data['locality_name'] == "Санкт-Петербург"]
    .dropna(subset=['city_center_km'])
    .sort_values(by='city_center_km')
)
print('Средняя стоимость км: ', (data_spb['last_price'].iloc[0] - data_spb['last_price'].iloc[-1]) / data_spb['city_center_km'].iloc[-1])

data_spb.plot(x='city_center_km', y='price_for_1m', style='o', alpha=0.2)
data_spb['city_center_km'].corr(data_spb['price_for_1m'])
```

**Вывод 4.5** Согласно расчетам, средняя стоимость одного км от центра составляет 82тыс руб. Также был построен график зависимости, где видно, что стоимсоть квартиры уменьшается при отдалении от центра
